## 2. Basic IO

### 2.A. Loading An Image

The function `imread()` can be used to load an image from a file:

```python
img = cv2.imread('path/to/img.ext', FLAG)
```
- First argument is the path to the file
- Second argument( optional) could be:

    - `cv2.IMREAD_COLOR` (or 1): Loads a color image. Any transparency of image will be neglected. It is the default flag.
    - `cv2.IMREAD_GRAYSCALE` (or 0): Loads image in grayscale mode
    - `cv2.IMREAD_UNCHANGED` (or -1): Loads image as such including alpha channel
    - ...

- Images loaded by OpenCV are Numpy arrays

### 2.B Saving An Image

Saving a numpy array (2D or 3D) as an image to the disk can be done using `imwrite`:

```python
cv2.imwrite('path/to/img.ext', img)
```

Most image extensions can be used: jpg, png, bmp...

### 3.A. Creating A Grayscale Image

Create a grayscale image of size 640x480, with a black background and a white 100x100 square in the center

````python
%matplotlib inline 

#import matplotlib
#matplotlib.use('TKAgg')
import matplotlib.pyplot as plt

import numpy as np
import cv2

# create a black image 640x480
image_dimensions_bg = (480, 640)
image_dimensions_fg = (100, 100)

img_black = np.zeros(image_dimensions_bg, dtype=np.uint8)

img_white_square = np.zeros(image_dimensions_bg, dtype=np.uint8)

top_margin = (image_dimensions_bg[0] - image_dimensions_fg[0]) / 2 - 1
bottom_margin = (image_dimensions_bg[0] - image_dimensions_fg[0]) / 2 + image_dimensions_fg[0]
left_margin = (image_dimensions_bg[1] - image_dimensions_fg[1]) / 2 - 1
right_margin = (image_dimensions_bg[1] - image_dimensions_fg[1]) / 2 + image_dimensions_fg[1]

for row in range(img_white_square.shape[0]):
    for col in range(img_white_square.shape[1]):
        if row > top_margin and row < bottom_margin and col > left_margin and col < right_margin:
            img_white_square[row,col] = 255

plt.imshow(cv2.cvtColor(img_black, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img_white_square, cv2.COLOR_BGR2RGB))
plt.show()
            
# write both the the disk
# img_black --> 'black.jpg' 
# img_white_square -->'white-square.jpg'
cv2.imwrite('black.jpg', img_black)
cv2.imwrite('white-square.jpg', img_white_square)
````

### 3.B. Creating A Color Image

- Create a color image of size 640x480, with a white background and a red 100x100 square in the center

- Create a color image of size 640x480, with random pixels

**Beware:** the order of the colors in the third dimension when writing is BGR, not RGB.

````python
%matplotlib inline 

#import matplotlib
#matplotlib.use('TKAgg')
import matplotlib.pyplot as plt

import numpy as np
import cv2

# create a white RGB-image 640x480
image_dimensions_bg = (480, 640, 3)
image_dimensions_fg = (100, 100, 3)

img_white = np.ones(image_dimensions_bg, dtype=np.uint8) * 255

img_red_square = np.ones(image_dimensions_bg, dtype=np.uint8) * 255

top_margin = (image_dimensions_bg[0] - image_dimensions_fg[0]) / 2 - 1
bottom_margin = (image_dimensions_bg[0] - image_dimensions_fg[0]) / 2 + image_dimensions_fg[0]
left_margin = (image_dimensions_bg[1] - image_dimensions_fg[1]) / 2 - 1
right_margin = (image_dimensions_bg[1] - image_dimensions_fg[1]) / 2 + image_dimensions_fg[1]

for row in range(img_red_square.shape[0]):
    for col in range(img_red_square.shape[1]):
        if row > top_margin and row < bottom_margin and col > left_margin and col < right_margin:
            img_red_square[row,col,0:2] = 0

img_noise = np.random.randint(0, 256, image_dimensions_bg, dtype=np.uint8)

plt.imshow(cv2.cvtColor(img_white, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img_red_square, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img_noise, cv2.COLOR_BGR2RGB))
plt.show()
            
# write all three to the the disk
# img_white --> 'white.jpg', 
# img_red_square --> 'red_square.jpg'
# img_noise --> 'noise.jpg'    
cv2.imwrite('white.jpg', img_white)
cv2.imwrite('red_square.jpg', img_red_square)
cv2.imwrite('noise.jpg', img_noise)
````

## 5. Point Operation: Thresholding

- The thresholding operation is to create a *destination* image $g$ from a *source* image $f$ by replacing each pixel of the original image $f$ with 255 if the pixel intensity is above a threshold $T$, and by 0 otherwise.

$$
    g(i, j) = \begin{cases}
                255 \quad \text{ if } f(i, j) > T \\
                0   \quad \text{ otherwise.}
            \end{cases}
$$

- Apply a threshold $T=127$ on the grayscale image of Lena `img_gray` and save the result to `lena-thresholded.jpg`. You can either:
    + Iterate over the two coordinates and set the pixel intensity of the target image/array:

    ```python
    for i in range(img_gray.shape[0]):
        for j in range(img_gray.shape[1]):
            thresholded[i, j] = ...
    ```
    
    + Or you use boolean arrays for sub-indexing:

    ```python
    thresholded[img_gray > T] = ...
    ```

````python
%matplotlib inline 

#import matplotlib
#matplotlib.use('TKAgg')
import matplotlib.pyplot as plt

import numpy as np
import cv2

img_gray = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.show()

# Threshold the lena image
T = 127

thresholded = np.copy(img_gray)
for row in range(thresholded.shape[0]):
    for col in range(thresholded.shape[1]):
        if thresholded[row,col] > T:
            thresholded[row,col] = 255
        else:
            thresholded[row,col] = 0
plt.imshow(cv2.cvtColor(thresholded, cv2.COLOR_BGR2RGB))
plt.show()

# write the result to the disk
cv2.imwrite('lena-thresholded.jpg', thresholded)

# or using sub-indexing
thresholded2 = np.copy(img_gray)
thresholded2[thresholded2 > T] = 255
thresholded2[thresholded2 <= T] = 0
plt.imshow(cv2.cvtColor(thresholded2, cv2.COLOR_BGR2RGB))
plt.show()

# write the result to the disk
cv2.imwrite('lena-thresholded2.jpg', thresholded2)
````

- Modify the thresholding procedure so that above-threshold pixels are converted to red, not white

## 6. Global Operation: Histogram Equalization

- Build and plot an histogram $h(I)$ of pixel intensities in the grayscale lena

- Build and plot the corresponding cumulative histogram $c(I)$
 
$$ c(I) = \frac{1}{N} \sum_{i=0}^I h(i) = c(I-1) + \frac{1}{N} h(I) $$

<table>
<tr>
<td>
<img src="img/lena-histo.png" alt="img/lena-histo.png\" width="300"/>
</td>
<td>
<img src="img/lena-cumhisto.png" alt="img/lena-cumhisto.png" width="300"/>
</td>
</tr>
</table>

````python
%matplotlib inline 

#import matplotlib
#matplotlib.use('TKAgg')
import matplotlib.pyplot as plt

import numpy as np
import cv2

img_gray = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.show()

# Make a histogram of pixel intensities
maxval = np.iinfo(img_gray.dtype).max
hist = np.zeros(maxval + 1)
for row in range(img_gray.shape[0]):
    for col in range(img_gray.shape[1]):
        hist[img_gray[row,col]] += 1

# Compute the cumulative histogram
cum_hist = np.zeros(maxval + 1)
previous_sum = 0
for idx in range(hist.shape[0]):
    cum_hist[idx] = previous_sum + hist[idx]
    previous_sum = cum_hist[idx]
cum_hist /= cum_hist.max()

# plot the results
plt.figure()
plt.subplot(1,2,1)    
plt.plot(hist)
plt.xlim((0, maxval))
plt.title('Histogram')
plt.xlabel('Pixel intensity')
plt.ylabel('Absolute number of pixels')

plt.subplot(1,2,2)    
plt.plot(cum_hist)
plt.xlim((0, maxval))
plt.title('Cumulative histogram')
plt.xlabel('Pixel intensity')
plt.ylabel('Relative number of pixels')

plt.show()
````

- Equalize the grayscale lena and save it to disk. Compare.

$$ g(i, j) = 255 \times c(f(i, j)) $$

````python
# Equalize the histogram
equalized = np.zeros(img_gray.shape, dtype=np.uint8)

for row in range(img_gray.shape[0]):
    for col in range(img_gray.shape[1]):
        equalized[row,col] = np.rint(255 * cum_hist[img_gray[row,col]])

plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(equalized, cv2.COLOR_BGR2RGB))
plt.show()

# write the result to the disk
cv2.imwrite('lena-equalized.jpg', equalized)
````

## 7. Local Operation: Locally Adaptive Histogram Equalization

- The problem with histogram equalization is that it uses **global** information about the image

- **Locally adaptive histogram equalization** (CLAHE) performs histogram equalization **locally** around each pixel.

<img src="img/clahe.png" alt="img/clahe.png" width="300"/>
<!-- ![](img/clahe.png) -->

$$ c(f(i, j)) = d_1(i, j) c_1(f(i, j)) + d_2(i, j) c_2(f(i, j)) + d_3(i, j) c_3(f(i, j)) + d_4(i, j) c_4(f(i, j)) $$

- OpenCV allows to apply directly this algorithm on an image:

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl = clahe.apply(img)
```

- Apply the CLAHE algorihm on the images lena and sudoku

````python
%matplotlib inline 

#import matplotlib
#matplotlib.use('TKAgg')
import matplotlib.pyplot as plt

import numpy as np
import cv2

# CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

# apply to Lena (img_gray) and save
img_gray = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.show()

clahe_lena = clahe.apply(img_gray)
plt.imshow(cv2.cvtColor(clahe_lena, cv2.COLOR_BGR2RGB))
plt.show()
cv2.imwrite('lena-clahe.jpg', clahe_lena)

# apply to sudoku (sudoku_gray) and save
sudoku_gray = cv2.imread('sudoku.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(sudoku_gray, cv2.COLOR_BGR2RGB))
plt.show()

clahe_sudoku = clahe.apply(sudoku_gray)
plt.imshow(cv2.cvtColor(clahe_sudoku, cv2.COLOR_BGR2RGB))
plt.show()
cv2.imwrite('sudoku-clahe.jpg', clahe_sudoku)
````

## 8. Color Filters

We want to filter out the feathers on Lena's hat. Their color has a hue between 120 and 160.

- Create a color image where only the feathers are colored as in the original, the rest being gray.

The algorithm is basically:

1. Convert the color image to HSV.

2. Convert the gray image to BGR to store the result.

3. For all pixels:    
    + If the hue is between 120 and 160, write the original BGR color vector to the result image.
    

- As a first try, make a double for-loop over the pixels

````python
%matplotlib inline 

#import matplotlib
#matplotlib.use('TKAgg')
import matplotlib.pyplot as plt

import numpy as np
import cv2

img_color = cv2.imread('lena.jpg', cv2.IMREAD_COLOR)
hsv = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)
gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

result = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

hue_min = 120
hue_max = 160

for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        if hsv[row,col,0] in range(hue_min, hue_max + 1):
            result[row,col] = img_color[row,col]
            
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
````

- As a second try, try to think about what this function does:

hue = hsv[:, :, 0]

````python
%matplotlib inline 

#import matplotlib
#matplotlib.use('TKAgg')
import matplotlib.pyplot as plt

import numpy as np
import cv2

img_color = cv2.imread('lena.jpg', cv2.IMREAD_COLOR)
hsv = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)
gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

result = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

hue_min = 120
hue_max = 160

hue = hsv[:,:,0]
locations_with_relevant_hue = np.logical_and(hue >= hue_min, hue <= hue_max)
result[locations_with_relevant_hue] = img_color[locations_with_relevant_hue]

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
````

````python
%matplotlib inline 

#import matplotlib
#matplotlib.use('TKAgg')
import matplotlib.pyplot as plt

import numpy as np
import cv2

# Filter the violet color
def hue_filter(img_color, hue_min, hue_max):
    img_hsv = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)
    img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
    result = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
    
    hue = img_hsv[:,:,0]
    locations_with_relevant_hue = np.logical_and(hue >= hue_min, hue <= hue_max)
    result[locations_with_relevant_hue] = img_color[locations_with_relevant_hue]
   
    return result

img_color = cv2.imread('lena.jpg', cv2.IMREAD_COLOR)

hue_min = 120
hue_max = 160

# call the function
violet_filtered = hue_filter(img_color, hue_min, hue_max)

plt.imshow(cv2.cvtColor(violet_filtered, cv2.COLOR_BGR2RGB))
plt.show()

# write to the disk
# compare color of feathers of lena-violet-filtered.jpg and lena-violet-filtered.bmp
# against the original lena.jpg, what do you find?
cv2.imwrite('lena-violet-filtered.jpg', violet_filtered)
cv2.imwrite('lena-violet-filtered.bmp', violet_filtered)
````

Saving a video is just as easy. Copy the code to a script `vwrite.py` and run it:


```python
# vwrite.py: writing a video

import cv2
import numpy as np

# create VideoWriter object
out = cv2.VideoWriter( 
    filename='video-square.ogv', 
    fourcc=cv2.VideoWriter_fourcc(*'THEO'), 
    fps=30.0, 
    frameSize=(560,320)
)


for f in range(90):
    frame = np.zeros((320,560), np.uint8)
    frame[160 - 50 + 2*f: 160 + 50 + 2*f, 
         280 - 50 + 2*f: 280 + 50 + 2*f] = 255
    out.write(cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR))

out.release()
```

- You only need to choose the right codec for the `fourcc` argument.

    + `'H264'` for mp4.
    + `'XVID'` for avi.
    + `'THEO'` for theora. 

- **Only Theora works in B202**
- See <http://www.fourcc.org>.

### 10.A. Processing Videos

- Read the given video `video.ogv` and save the grayscale version in `video-gray.ogv`.

- Apply a color filter on each frame and filter the blue color (hue between 100 and 130).

- Compare the time needed for this last conversion when using the double for-loops, and the `numpy.bitwise_and` method.


````python
# convert video to grayscale

import cv2

cap = cv2.VideoCapture('video.ogv')
inputFPS = int(cap.get(cv2.CAP_PROP_FPS))
inputFrameSize = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

# create VideoWriter object
out = cv2.VideoWriter( 
    filename='video-gray.mkv', 
    fourcc=cv2.VideoWriter_fourcc(*'VP80'), 
    fps=inputFPS,
    frameSize=inputFrameSize
)

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        # make the frame grayscale
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # since write() expects bgr images, we have to convert it back to bgr
        frame = cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR)
        out.write(frame)
    else:
        break

out.release()
cap.release()
````

````python
# color filter benchmark

import cv2
import numpy as np
import timeit

def hue_filter(img_color, hue_min, hue_max, processing_method):
    img_hsv = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)
    img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
    result = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
    
    if processing_method == 'LOGICAL_AND':
        hue = img_hsv[:,:,0]
        locations_with_relevant_hue = np.logical_and(hue >= hue_min, hue <= hue_max)
        result[locations_with_relevant_hue] = img_color[locations_with_relevant_hue]
    elif processing_method == 'DOUBLE_FOR_LOOP':
        for row in range(result.shape[0]):
            for col in range(result.shape[1]):
                if img_hsv[row,col,0] in range(hue_min, hue_max + 1):
                    result[row,col] = img_color[row,col]
    else:
        result = None
   
    return result

hue_min = 100
hue_max = 130

###################################

cap = cv2.VideoCapture('video.ogv')

tic = timeit.default_timer()

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        # filter the violet color
        frame = hue_filter(frame, hue_min, hue_max, 'LOGICAL_AND')
        #cv2.imshow('frame', frame)
    else:
        break
    k = cv2.waitKey(5) & 0xFF # Sleep 5 ms
    if k == 27: # Press ESC to close
        break

toc = timeit.default_timer()
print("Execution using logical_and implementation took: " + str(toc-tic) + "s") 
cap.release()

###################################

cap = cv2.VideoCapture('video.ogv')

tic = timeit.default_timer()

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        # filter the violet color
        frame = hue_filter(frame, hue_min, hue_max, 'DOUBLE_FOR_LOOP')
        #cv2.imshow('frame', frame)
    else:
        break
    k = cv2.waitKey(5) & 0xFF # Sleep 5 ms
    if k == 27: # Press ESC to close
        break

toc = timeit.default_timer()
print("Execution using double for loop implementation took: " + str(toc-tic) + "s") 
cap.release()

#cv2.destroyAllWindows()
````
